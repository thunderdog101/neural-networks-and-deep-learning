```
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.style as style
```

```
# Read data
edu_data = pd.read_csv('http://www.randalolson.com/wp-content/uploads/percent-bachelors-degrees-women-usa.csv')
```

```
edu_data.head()
```

```
edu_data.loc[0]
```

```
# Finding subset of degrees we wish to plot
under_20_pct = edu_data.loc[0, edu_data.loc[0] < 20].index
under_20_pct
```

```
# Basic first plot
edu_data.plot(x='Year', y=under_20_pct, figsize=(15, 10))
```

```
style.available
```

```
style.use('fivethirtyeight')
```

```
# Color blind friendly colors

# Colorblind-friendly colors
colors = [[0,0,0], [230/255,159/255,0], [86/255,180/255,233/255], [0,158/255,115/255],
          [213/255,94/255,0], [0,114/255,178/255]]
```

```
edu_plot = edu_data.plot(x='Year',  y=under_20_pct, figsize=(15, 10), color=colors)

# Bold horizontal line at y=0
edu_plot.axhline(y=0, alpha=0.7, linewidth = 1.5, color='black')

# Tick labels
edu_plot.tick_params(axis='both', which='major', labelsize=18)

# Set y ticks
edu_plot.set_yticklabels(labels = [-10, '0   ', '10   ', '20   ', '30   ', '40   ', '50%'])

# Add an extra vertical line by tweaking range of the x-axis
edu_plot.set_xlim(left=1969, right=2011)

# Adding a signature bar
edu_plot.xaxis.label.set_visible(False)

edu_plot.text(x=1966,
              y=-7,
              s="""© CS06 Data Viz with Python             """
                + """                                                        """
                + """Source: National Center for Education Statistics""",
              fontsize=17,
              color='#f0f0f0',
              backgroundcolor='grey'
              )

# Adding a title and a subtitle
edu_plot.text(x = 1966.65, y = 62.7, s = "The gender gap is transitory - even for extreme cases",
              fontsize = 26, weight = 'bold', alpha = .75)
edu_plot.text(x = 1966.65, y = 57,
              s = 'Percentage of Bachelors conferred to women from 1970 to 2011 in the US for\nextreme cases where the percentage was less than 20% in 1970',
              fontsize = 19, alpha = .85)

plt.show()
```
